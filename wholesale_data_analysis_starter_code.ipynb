## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df['subtotal'].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda x: 7 * x if x > 50 else 10 * x)
df['shipping_price'].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = (df['subtotal'] + df['shipping_price']) * (1.0925)
df['total_price'] = df['total_price'].round(2)
df['total_price'].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df['line_cost'].head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['total_price'] - df['line_cost']
df['profit'].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
df['quantity_spend'] = (df['total_price']/df['qty']).round(2)
id_33615 = float(df[df['client_id'] == 33615]['quantity_spend'].sum())
id_66037 = float(df[df['client_id'] == 66037]['quantity_spend'].sum().round(2))
id_46820 = float(df[df['client_id'] == 46820]['quantity_spend'].sum().round(2))
id_24741 = float(df[df['client_id'] == 24741]['quantity_spend'].sum())
id_38378 = float(df[df['client_id'] == 38378]['quantity_spend'].sum())
print(f"Order ID 33615 quantity spend:", id_33615)
print(f"Order ID 66037 quantity spend:", id_66037)
print(f"Order ID 46820 quantity spend:", id_46820)
print(f"Order ID 24741 quantity spend:", id_24741)
print(f"Order ID 38378 quantity spend:", id_38378)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_df = df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'profit': 'sum'
}).reset_index()
summary_df = summary_df.sort_values(by='total_price', ascending=False)
top_5_clients = summary_df.head(5)
top_5_clients.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Price', 'Total Revenue', 'Total Profit']
print(top_5_clients)
```

```python
# Format the data and rename the columns to names suitable for presentation.
df = df.rename(columns={'first': 'first_name'})
df = df.rename(columns={'last': 'last_name'})
df = df.rename(columns={'qty': 'quantity'})
# Define the money columns. 
money_columns = ['line_cost', 'subtotal', 'shipping_price', 'total_price', 'profit', 'quantity_spend']

# Define a function that converts a dollar amount to millions.
def to_millions(amount):
    """Convert a dollar amount to millions."""
    return round(amount / 1_000_000, 6)

# Apply the currency_format_millions function to only the money columns. 
df[money_columns] = df[money_columns].applymap(to_millions)

# Rename the columns to reflect the change in the money format. 
df = df.rename(columns={'line_cost': 'line_cost_in_millions'})
df = df.rename(columns={'subtotal': 'subtotal_in_millions'})
df = df.rename(columns={'shipping_price': 'shipping_price_in_millions'})
df = df.rename(columns={'total_price': 'total_price_in_millions'})
df = df.rename(columns={'profit': 'profit_in_millions'})
df = df.rename(columns={'quantity_spend': 'quantity_spend_in_millions'})
df.head()

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df = df.sort_values(by='profit_in_millions', ascending=False)
sorted_df
```
